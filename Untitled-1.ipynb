```python
import clip
import torch
from PIL import Image
```

```python
import sys
print(sys.executable)
```

```python
labels = ["a dog", "a cat", "a truck", "a sunset", "a person", "a pizza", "a painting", "a diagram"]
text = clip.tokenize(labels)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

image = preprocess(Image.open("C:/Users/Home/Documents/images1.png")).unsqueeze(0).to(device)
```

```python
with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)

    logits_per_image, _ = model(image, text)
    probs = logits_per_image.softmax(dim=-1).cpu().numpy()
```

```python
import numpy as np

best_idx = np.argmax(probs)
print("üîç CLIP thinks this is:", labels[best_idx])
print("‚úÖ Confidence:", probs[0][best_idx])
```
